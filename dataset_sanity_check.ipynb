```python
import IPython.display as ipd
```

```python
SAMPLE_RATE = 16_000
```

```python
from datasets.dataset import ChunkDataset
from datasets.configs import BaseDatasetConfig

config = BaseDatasetConfig(id="gtzan")

ds = ChunkDataset(
    parquet_file_path=config.parquet,
    h5_file_path = config.hdf5,
    split="train",
)
```

```python
sample = ds[0]
audio = sample["waveform"]
display(ipd.Audio(audio, rate=SAMPLE_RATE, autoplay=False))
```
